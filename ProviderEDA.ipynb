## **HISTOGRAM**

```python
df.Provider_State.value_counts().nlargest(40).plot(kind='bar', figsize=(10,5))
plt.title('Number of Providers by State')
plt.ylabel('Number of Providers')
plt.xlabel('State')
```

```python
df.Provider_Type.value_counts().plot(kind='bar', figsize=(10,5))
plt.title('Number of Providers by Type')
plt.ylabel('Number of Providers')
plt.xlabel('Type')
```

```python
df.Primary_Care_ProviderPCP.value_counts().plot(kind='bar', figsize=(10,5))
plt.title('Number of Providers by Speciality')
plt.ylabel('Number of Providers')
plt.xlabel('Speciality')
```

## **HEAT MAP**

```python
data = df.drop(['Federal Provider Number','Provider City','Provider Zip Code', 'Provider Name', 'Provider Address', 'Provider Phone Number','Provider County Name', 'Patient satisfaction rate(out of 10)'], axis=1)
```

```python
corr = data.corr()
ax = sns.heatmap(
    corr, 
    vmin=-1, vmax=1, center=0,
    cmap=sns.diverging_palette(20, 220, n=200),
    square=True
)
ax.set_xticklabels(
    ax.get_xticklabels(),
    horizontalalignment='right'
);
```
